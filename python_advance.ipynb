```python
class Person :
  def __init__(self, nama, usia, saldo):
    self.nama =nama
    self.usia = usia
    self.__saldo = saldo
  def deposit(self, amount):
    self.__saldo += amount
  def tarik(self, amount):
    if self.__saldo >= amount :
      self.__saldo -= amount
    else:
      print("Saldo tidak mencukupi")
  def cek_saldo(self):
    return self.__saldo

person1 = Person('Egg', 24, 100)
person1.deposit(50)
person1.tarik(51)
person1.tarik(100)


print(f'Sisa saldo : {person1.cek_saldo()}')
```

## Number 4

```python
class Person :
  def __init__(self, nama, usia):
    self.nama = nama
    self.usia = usia
  def tabungan(self):
    print('Person menabung sejumlah tertentu')

class Mahasiswa(Person):
  def tabungan(self, tabungan):
    print(f'Mahasiswa {self.nama} menabung {tabungan}')

class Pegawai(Person):
  def tabungan(self, jumlah,  gaji):
    print(f"Pegawai {self.nama} menabung {jumlah} dari gaji {gaji}")

class Dosen(Person):
  def tabungan(self,jumlah, mata_kuliah):
    print(f"Dosen {self.nama} menabung {jumlah} dari honor {mata_kuliah}")

person = Person('egg', 24)
mahasiswa = Mahasiswa('egg', 24)
pegawai = Pegawai('ratna', 21)
dosen = Dosen('erik', 27 )

person.tabungan()
mahasiswa.tabungan(100000)
pegawai.tabungan(12000, 500000)
dosen.tabungan(1200, 'PCD')
```

```python
from abc import ABC, abstractmethod

class TabunganInterface(ABC):
    @abstractmethod
    def tabung(self, amount):
        pass

class Mahasiswa(TabunganInterface):
    def __init__(nama, self, saldo):
        self.nama = nama
        self.saldo = saldo
    def tabung(self, amount):
        self.saldo += amount
        print(f'Nama : {self.nama} Tabung : {amount} Saldo :{self.saldo}')

class Pegawai(TabunganInterface):
    def __init__(nama, self, saldo):
        self.nama = nama
        self.saldo = saldo
    def tabung(self, amount):
        self.saldo += amount
        print(f'Nama : {self.nama} Tabung : {amount} Saldo :{self.saldo}')

class Dosen(TabunganInterface):
    def __init__(self,nama, saldo):
        self.nama = nama
        self.saldo = saldo
    def tabung(self, amount):
        self.saldo += amount
        print(f'Nama : {self.nama}, Tabung : {amount}, Saldo :{self.saldo}')


dosen1 = Dosen('Egg',12000)
dosen1.tabung(100)
dosen1.tabung(10)

```
